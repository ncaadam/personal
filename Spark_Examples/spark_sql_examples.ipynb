#Playing with Spark SQL

## Reading a schema from a flat file
In this section we read data in from a text file and infer a schema, then perform a SQL query on the data structure

### Set up Spark

```python
from pyspark import SparkContext, SparkConf
```

```python
# Configure Spark Settings
conf=SparkConf()
#conf.set("spark.executor.memory", "1g")
#conf.set("spark.cores.max", "2")
conf.setAppName("My App")

## Initialize SparkContext
sc = SparkContext('local[*]', conf=conf)
```

```python
print sc
```

```python
# sc is an existing SparkContext.
from pyspark.sql import *
sqlContext = SQLContext(sc)
```

### Read in the flat file and set up the schema

```python
!head /Users/sfullman001/Documents/code/ipython_scrap/data/invoice_lines.txt
```

```python
# Load a text file
lines = sc.textFile("/Users/sfullman001/Documents/code/ipython_scrap/data/invoice_lines_short.txt")
```

```python
lines = lines.map(lambda l: l.split("|"))
header = lines.first()
parts = lines.filter(lambda line: True if line != header else False)
```

```python
header
```

```python
parts.take(1)
```

```python
parts.count()
```

```python
fields = [StructField(field_name, StringType(), True) for field_name in header]
schema = StructType(fields)
```

```python
schema
```

```python
schemaInvoice = sqlContext.applySchema(parts, schema)
schemaInvoice.registerTempTable("parts")
```

```python
schemaInvoice.printSchema()
```

### Do some queries

```python
results = sqlContext.sql("SELECT * FROM parts WHERE price <> 0")
```

```python
results.count()
```

```python
results.take(5)
```

##Use Spark SQL on data from a JSON file
In this section we read in some data from a database that contains information on companies from Crunchbase.  The interesting thing is that the schema is not consistent across the data, with some documents missing information

###Read in some data

```python
!head /Users/sfullman001/Documents/code/ipython_scrap/data/orgs.json
```

```python
companies = sqlContext.jsonFile('/Users/sfullman001/Documents/code/ipython_scrap/data/orgs.json')
```

```python
companies.count()
```

###Register a table and print the schema

```python
companies.registerTempTable("companies")
```

```python
companies.printSchema()
```

```python
scotts = sqlContext.sql("SELECT * FROM companies WHERE data.data.relationships.current_team.items[0].first_name LIKE \'%Scott%\'")
```

```python
scotts.count()
```

```python
scotts.take(1)
```
