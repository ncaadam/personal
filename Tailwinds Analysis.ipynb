```python
import graphlab as gl
import graphlab.aggregate as agg
import datetime
```

```python
ac_lookup = gl.SFrame.read_csv('../../Analytics/Tailwinds/AC Lookup.csv',verbose=False)
```

```python
ac_dict = {}
for row in ac_lookup:
    ac_dict[int(row['Code'])] = str(row['Description'])
```

```python
fuel_prices = gl.SFrame.read_csv('../../Analytics/Tailwinds/fuel_by_month - 90-15.csv',verbose=False)
```

# Fuel / MX Cost

```python
P52 = gl.SFrame.read_csv('../../Analytics/Tailwinds/924143566_T_F41SCHEDULE_P52/924143566_T_F41SCHEDULE_P52_All_All.csv',verbose=False)
```

```python
P52.remove_column('X64')
```

```python
for col in P52.column_names():
    P52.rename({col:col[1:-1]})
```

```python
p_data = P52.select_columns(['REGION','YEAR','QUARTER','CARRIER_GROUP','FUEL_FLY_OPS','AIRCRAFT_TYPE','AIR_DAYS_ASSIGN',
                           'TOTAL_AIR_HOURS','TOT_AIR_OP_EXPENSES','PILOT_FLY_OPS','TOT_FLY_OPS','AP_MT_BURDEN',
                           'TOT_DIR_MAINT'])
```

```python
p_filtered = p_data.filter_by('D','REGION').filter_by(3,'CARRIER_GROUP').filter_by(range(2012,2015),'YEAR')
```

```python
p_filtered['AIRCRAFT_NAME'] = p_filtered['AIRCRAFT_TYPE'].apply(lambda x: ac_dict[int(x)])
```

```python
p_filtered['AIR_DAYS_ASSIGN'] = p_filtered['AIR_DAYS_ASSIGN'].astype(float,True)
p_filtered['TOTAL_AIR_HOURS'] = p_filtered['TOTAL_AIR_HOURS'].astype(float,True)
p_filtered['PILOT_FLY_OPS'] = p_filtered['PILOT_FLY_OPS'].astype(float,True)
```

```python
p_aggregate = p_filtered.groupby(['YEAR','QUARTER','AIRCRAFT_NAME'],
                 operations=[agg.SUM('FUEL_FLY_OPS'),agg.SUM('AIR_DAYS_ASSIGN'),agg.SUM('TOTAL_AIR_HOURS'),
                             agg.SUM('PILOT_FLY_OPS'),agg.SUM('TOT_FLY_OPS'),agg.SUM('AP_MT_BURDEN'),
                             agg.SUM('TOT_DIR_MAINT')])
```

# ASMs

```python
T2 = gl.SFrame.read_csv('../../Analytics/Tailwinds/924382816_T_SCHEDULE_T2/924382816_T_SCHEDULE_T2_All_All.csv',verbose = False)
```

```python
T2.remove_column('X32')
```

```python
for col in T2.column_names():
    T2.rename({col:col[1:-1]})
```

```python
t_data = T2.select_columns(['QUARTER','YEAR','CARRIER_REGION','CARRIER_GROUP_NEW','AIRCRAFT_TYPE','AVL_SEAT_MILES_320'])
```

```python
t_filtered = t_data.filter_by('D','CARRIER_REGION').filter_by(3,'CARRIER_GROUP_NEW').filter_by(range(2012,2015),'YEAR')
```

```python
t_filtered['AIRCRAFT_NAME'] = t_filtered['AIRCRAFT_TYPE'].apply(lambda x: ac_dict[int(x)])
```

```python
t_filtered['AVL_SEAT_MILES_320'] = t_filtered['AVL_SEAT_MILES_320'].astype(float,True)
```

```python
t_agg = t_filtered.groupby(['QUARTER','YEAR','AIRCRAFT_NAME'],agg.SUM('AVL_SEAT_MILES_320'))
```

```python
joined = t_agg.join(p_aggregate,on=['YEAR','QUARTER','AIRCRAFT_NAME'])
```

```python
col_list = ['Sum of FUEL_FLY_OPS','Sum of AIR_DAYS_ASSIGN','Sum of TOTAL_AIR_HOURS','Sum of PILOT_FLY_OPS',
            'Sum of TOT_FLY_OPS','Sum of AP_MT_BURDEN','Sum of TOT_DIR_MAINT']
joined_prop = joined
for col in col_list:
    joined_prop[col] = joined_prop[col]*1000
```

```python
joined_prop['UTIL'] = joined_prop['Sum of TOTAL_AIR_HOURS']/joined_prop['Sum of AIR_DAYS_ASSIGN']
```

```python
joined_prop['CASM'] = joined_prop['Sum of AVL_SEAT_MILES_320']/(joined_prop['Sum of FUEL_FLY_OPS'] + joined_prop['Sum of TOT_DIR_MAINT'])
```

```python
qs = {'Jan':1,'Feb':1,'Mar':1,'Apr':2,'May':2,'Jun':2,'Jul':3,'Aug':3,'Sep':3,'Oct':4,'Nov':4,'Dec':4}
```

```python
fuel_prices['YEAR'] = 0
fuel_prices['QUARTER'] = 0
```

```python
fuel_prices['YEAR'] = fuel_prices['Month'].apply(lambda x: int('19'+ x[-2:]) if int(x[-2:]) > 16 else int('20' + x[-2:]))
```

```python
fuel_prices['QUARTER'] = fuel_prices['Month'].apply(lambda x: qs[x[:3]])
```

```python
fuel_prices_agg = fuel_prices.groupby(['YEAR','QUARTER'],{'Quarterly Fuel Price':agg.AVG('Price')})
```

```python
joined_prop_fuel = joined_prop.join(fuel_prices_agg,on=['YEAR','QUARTER'],how='inner')
```

```python
final = joined_prop_fuel
```

```python
final.save('../../Analytics/Tailwinds/aggregate.csv')
```
