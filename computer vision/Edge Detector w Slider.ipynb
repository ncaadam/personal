```python
from SimpleCV import *
disp = Display(displaytype = 'notebook')
```

#### This cell creates the core function of our script. Normally it would go in a function with an image path as a parameter. However, for the sake simplicity, it is just two cells. This script couldn't be broken up very well as it has a GUI. Therefore, execute both cells and it will save the image at the bottom.  <br><br>The function creates a window and displays the image. The image's edges are being detected based on the threshold values that it is given. The threshold values can be changed real-time via the sliders at the top of the window.  <br><br>Once you're finished adjusting the sliders, press ESC to close the window and save the output to the ipython notebook.

```python
#loading images
img = cv2.imread('../images/coke_test.jpg')

#get the height and width
w,h = img.shape[:2]

#scale down the image
w,h = int(w*.2),int(h*.2)
img = cv2.resize(img,(h,w))
    
#create image window and start window thread
cv2.namedWindow('image')
cv2.startWindowThread()

#initialize parameters
th1 = 0
th2 = 0

#display initial edge detection image
edges = cv2.Canny(img,th1,th2)
cv2.imshow('image',edges)

#create the function that triggers when slider 1's value is changed
def update1(value):
    th2 = cv2.getTrackbarPos('Threshold 2','image')
    edges = cv2.Canny(img,value,th2)
    cv2.imshow('image',edges)

#create the function that triggers when slider 2's value is changed
def update2(value):
    th1 = cv2.getTrackbarPos('Threshold 1','image')
    edges = cv2.Canny(img,th1,value)
    cv2.imshow('image',edges)    

# create trackbars for threshold change
cv2.createTrackbar('Threshold 1','image',th1,255,update1)
cv2.createTrackbar('Threshold 2','image',th2,255,update2)

#press ESC to stop
while(1):
    k = cv2.waitKey(0) & 0xFF
    if k == 27:
        break
        
#set the variables to the trackbar positions
th1 = cv2.getTrackbarPos('Threshold 1','image')
th2 = cv2.getTrackbarPos('Threshold 2','image')
#create the final image
edges = cv2.Canny(img,th1,th2)
#properly destroy the window
cv2.destroyAllWindows()
cv2.waitKey(1)

#print the thresholds that we used for the image
print "Threshold 1:", th1
print "Threshold 2:", th2

#convert the edges image into a SimpleCV image (this helps display inline in ipython)
orig_scv = Image(edges,cv2image=True)
edges_scv = Image(img,cv2image=True)
#show the original next to the edges picture
orig_scv.sideBySide(edges_scv,side='bottom').save(disp)
```
