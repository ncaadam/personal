<center><h1>Edge Detector</h1></center>
<b>Description: </b>This script can be used to find the edges in an image. In this example, we're finding the edges of cans in a cooler. You can change the code in cell 18 to import your own picture.
<br><b>Packages Required: </b>
<ul>
<li><u>SimpleCV</u> - <i><b>Note:</b> SimpleCV requires OpenCV</i></li>
<li><u>NumPy</u></li></ul>

```python
import numpy as np
from SimpleCV import *
disp = Display(displaytype = 'notebook')
```

```python
def canny_sliders(path): 
    #loading images
    img = cv2.imread(path)
    
    #get the height and width
    w,h = img.shape[:2]
    
    #scale down the image
    w,h = int(w*.2),int(h*.2)
    img = cv2.resize(img,(h,w))
        
    #create image window and start window thread
    cv2.namedWindow('image')
    cv2.startWindowThread()
    
    def nothing(nothing):
        pass
    
    # create trackbars for threshold change
    cv2.createTrackbar('Threshold 1','image',0,255,nothing)
    cv2.createTrackbar('Threshold 2','image',0,255,nothing)
    
    while(1):
        # get current positions of our trackbars
        th1 = cv2.getTrackbarPos('Threshold 1','image')
        th2 = cv2.getTrackbarPos('Threshold 2','image')
        #apply canny 
        edges = cv2.Canny(img,th1,th2)
        #show the image in the window
        cv2.imshow('image',edges)
        #press ESC to stop
        k = cv2.waitKey(0) & 0xFF
        if k == 27:
            break
    
    #properly destroy the window
    cv2.destroyAllWindows()
    cv2.waitKey(1)
    
    #print the thresholds that we used for the image
    print "Threshold 1:", th1
    print "Threshold 2:", th2
    
    #return the pictures and the threshold values
    return img,edges
```

#### This cell creates the core function of our script: <i>canny_sliders()</i>. The function creates a window and displays the image. The image's edges are being detected based on the threshold values that it is given. The threshold values can be changed real-time via the sliders at the top of the window. <br><br>Once you're finished adjusting the sliders, press ESC to close the window and save the output to the ipython notebook.

```python
#run the function we just created
#to break the window, press esc
orig,edges = canny_sliders('coke_test.jpg')
#convert the edges image into a SimpleCV image (this helps display inline in ipython)
orig_scv = Image(edges,cv2image=True)
edges_scv = Image(orig,cv2image=True)
#show the original next to the edges picture
orig_scv.sideBySide(edges_scv,side='bottom').save(disp)
```
