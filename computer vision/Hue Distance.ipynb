```python
from SimpleCV import Image, Color, Display
import pylab as plt
disp = Display(displaytype = 'notebook')
```

```python
# Load an image from imgur.
img = Image("../Cooler Images/coke_test.jpg")
scaledImg = img.scale(img.width/3,img.height/3)
scaledImg = scaledImg.crop(0,0,scaledImg.width,scaledImg.height)
scaledImg.save(disp)

peaks = scaledImg.huePeaks()
hist = scaledImg.hueHistogram()
plt.plot(hist)

hdist = scaledImg.hueDistance(peaks[3][0])
binary = scaledImg.hueDistance(peaks[3][0]).invert().threshold(110)
hdist.save(disp)
binary.save(disp)
#print output
#output = scaledImg.binarize(130).invert().edges()
#feats = output.findKeypoints().draw()
#output.show()
```
