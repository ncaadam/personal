```python
from SimpleCV import *
```

```python
disp = Display(displaytype='notebook')
temp = Image("../5.png")
img = Image("../cokecantest.jpg")
#img = img.scale(.4)
#img = img.grayscale()
#temp = temp.scale(.4)
```

```python
img.clearLayers()
hue_dist = img.hueDistance(Color.RED).invert()
hue_dist.save(disp)
watershed = hue_dist.watershed(erode=1,dilate=1)
blobs = watershed.findBlobsFromWatershed(watershed)
blobs = blobs.reassignImage(img)
blobs[-1].draw(width = -1, color = Color.RED)

img.save(disp)
```
