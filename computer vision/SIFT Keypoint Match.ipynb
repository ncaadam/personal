<center><h1>SIFT Keypoint Match</h1></center>
<b>Description: </b>This script can be used to find feature matches between two images. In this example, we're trying to see if there is a Coke can in a picture. The top example shows how well the keypoint match can work. The bottom example shows the pitfalls of the keypoint match function. The second examples uses a template I made myself vs. cropping a coke can from the image we're analyzing.
<br><b>Packages Required: </b><ul><li><u>SimpleCV</u> - <i><b>Note:</b> SimpleCV requires OpenCV</i></li></ul>

```python
from SimpleCV import Image,Display
disp = Display(displaytype='notebook')
```

```python
#import the template
temp=Image("../images/cokecantest_cropped.jpg")
#import the image to compare to
img=Image("../images/cokecantest.jpg")
#run the SIFT Keypoint Match
keys = img.drawSIFTKeyPointMatch(temp,distance=10,num=15,width=2)
keys.save(disp)
```

```python
#import the template
temp = Image("../images/5.png")
temp = temp.scale(.2)
#import the test image
img = Image("../images/cokecantest.jpg")
keys = img.drawSIFTKeyPointMatch(temp,distance=10,num=15,width=2)
keys.save(disp)
```

```python
#try it with a rotated image
temp_rotated = temp.rotate90()
keys = img.drawSIFTKeyPointMatch(temp_rotated,distance=10,num=15,width=2)
keys.save(disp)
```
