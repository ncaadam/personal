```python
import numpy as np
import SimpleCV as scv
```

```python
img = scv.cv2.imread('../Cooler Images/coke_test.jpg')
gray = scv.cv2.imread('../Cooler Images/coke_test.jpg',0)

ret,thresh = scv.cv2.threshold(gray,175,255,1)

contours,h = scv.cv2.findContours(thresh,1,2)

for cnt in contours:
    approx = scv.cv2.approxPolyDP(cnt,.01*scv.cv2.arcLength(cnt,True),True)
    print len(approx)
    if len(approx)==5:
        print "pentagon"
        scv.cv2.drawContours(img,[cnt],0,255,-1)
    elif len(approx)==3:
        print "triangle"
        scv.cv2.drawContours(img,[cnt],0,(0,255,0),-1)
    elif len(approx)==4:
        print "square"
        scv.cv2.drawContours(img,[cnt],0,(0,0,255),-1)
    elif len(approx) == 9:
        print "half-circle"
        scv.cv2.drawContours(img,[cnt],0,(255,255,0),-1)
    elif len(approx) > 15:
        print "circle"
        scv.cv2.drawContours(img,[cnt],0,(0,255,255),-1)
        
newx,newy = img.shape[1]/3,img.shape[0]/3
newImage = scv.cv2.resize(img,(newx,newy))
scv.cv2.imshow('img',newImage)
scv.cv2.waitKey(0)
scv.cv2.destroyAllWindows()
```
