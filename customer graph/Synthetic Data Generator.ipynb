```python
# this is for the synthetic CRM data model
class Phone(EmbeddedDocument):
    number = IntField() #generate
    phone_type = StringField() #generate
    
class CustomerProfile(EmbeddedDocument):
    customer_id = StringField() #generate
    name = StringField() #twitter
    location = StringField() #twitter
    email = StringField() #generate
    phone = EmbeddedDocumentField(Phone) #generate
    twitter_handle = StringField() #twitter
    dob = DateTimeField() #generate
    gender = StringField() #generate

class CustomerBehavior(EmbeddedDocument):
    pk = ObjectIdField() #forced to add (doesn't show up in the DB?)
    mobile_app = BooleanField() #generate
    gift_card_balance = FloatField() #generate
    number_of_transactions = IntField() #generate

class Employment(EmbeddedDocument): # TODO: TBD
    linkedin_profile = ObjectIdField() 
    employer = StringField()
    job_title = StringField()
    salary_range_low = IntField()
    salary_range_mid = IntField()
    salary_range_high = IntField()

class CrmData(EmbeddedDocument):
    pk = ObjectIdField() #forced to add (doesn't show up in the DB?)
    customer_profile = EmbeddedDocumentField(CustomerProfile)
    customer_behavior = EmbeddedDocumentField(CustomerBehavior)

class CollectedData(EmbeddedDocument):
    twitter_profile = ObjectIdField() #twitter profile foreign key
    twitter_tweets = ObjectIdField() #tweets document foreign key
    employment = EmbeddedDocumentField(Employment) #TODO: update this to be an embedded doc for the matching algo
    
class MatchedData(EmbeddedDocument):
    sbi = ObjectIdField() #TODO: update this to be an embedded doc for the matching algo

class CustomerData(Document):
    crm_data = EmbeddedDocumentField(CrmData)
    collected_data = EmbeddedDocumentField(CollectedData)
    matched_data = EmbeddedDocumentField(MatchedData)
```

```python
def generate_customer_id():
    new_id = uuid.uuid4()
    return str(new_id)
    

def generate_email(customer):
    if customer.crm_data.customer_profile.name:
        new_name = str(customer.crm_data.customer_profile.name.encode('utf-8')).lower().replace(' ', '')
        new_email = new_name + '@email.com'
        return new_email
    else:
        #create generic email if there is no name
        pass


def generate_phone_number():
    number = '2125551212'
    return number


def generate_phone_type():
    phone_type = 'mobile'
    return phone_type


from random import randrange
base = datetime.datetime.today()
def generate_dob():
    days_param = randrange(365*13,365*60)
    new_dob = base - datetime.timedelta(days=days_param)
    return new_dob.strftime('%m/%d/%Y')


male_female_list = ['male', 'female']
def generate_gender():
    gender = random.choice(male_female_list)
    return gender


mobile_app_probability = ['True'] * 3 + ['False'] * 7
def generate_mobile_app():
    return bool(random.choice(mobile_app_probability))


def generate_gift_card_balance():
    return random.uniform(0, 74.32)


def generate_number_of_transactions():
    #TODO: update this based on the parameters we need
    return 0


def create_customer_object():
    customer = CustomerData(
        crm_data = CrmData(
            customer_profile = CustomerProfile(
                phone = Phone()), 
            customer_behavior = CustomerBehavior()),
        collected_data = CollectedData(
            employment = Employment()),
        matched_data = MatchedData())
    return customer
```

```python
#grab each of the collections you need for your queries and document insert
profiles_coll = db.twitter_profiles
tweets_coll = db.all_user_tweets
cust_coll = db.customer_data
```

```python
#grab all the user IDs for the users with tweets
#to test this with just a few records, just add ".limit(1)" or ".limit(5)" to the end of the query
user_ids = db.all_user_tweets.find({},{"tweets.user.id":1})
#total # of users for tracking
array_count = user_ids.count()
#iterator for tracking
i = 0
#creates logs.txt and opens it with write capabilities
#NOTE - this will overwrite any file named logs.txt in the directory
with open("logs.txt",'w') as logs:
    logs.write("START TIME: " + str(datetime.datetime.now()) + "\n")
    for tweet_id in user_ids:
        try:
            #set twitter user ID
            user_id = tweet_id['tweets'][0]['user']['id']
            #set foreign key to tweets
            tweet_fk = tweet_id['_id']
            #grab twitter profile via twitter user id
            #return name, location, handle
            user_profile = profiles_coll.find({"profile.id":user_id},{"profile.name":1,"profile.location":1,"profile.screen_name":1})
            #set name
            user_name = user_profile[0]['profile']['name']
            #set handle
            user_handle = user_profile[0]['profile']['screen_name']
            #set location
            user_location = user_profile[0]['profile']['location']
            #set foreign key to twitter profile
            profile_fk = user_profile[0]['_id']
            #assign all the values to the customer object
            customer = create_customer_object()
            customer.crm_data.customer_profile.name = user_name
            customer.crm_data.customer_profile.location = user_location
            customer.crm_data.customer_profile.twitter_handle = user_handle
            customer.crm_data.customer_profile.customer_id = generate_customer_id()
            customer.crm_data.customer_profile.email = generate_email(customer)
            customer.crm_data.customer_profile.phone.number = generate_phone_number()
            customer.crm_data.customer_profile.phone.phone_type = generate_phone_type()
            customer.crm_data.customer_profile.dob = generate_dob()
            customer.crm_data.customer_profile.gender = generate_gender()
            customer.crm_data.customer_behavior.mobile_app = generate_mobile_app()
            customer.crm_data.customer_behavior.gift_card_balance = generate_gift_card_balance()
            customer.crm_data.customer_behavior.number_of_transactions = generate_number_of_transactions()
            customer.collected_data.twitter_profile = profile_fk
            customer.collected_data.twitter_tweets = tweet_fk
            #save the document to the DB
            customer.save()
            #write to the log file
            logs.write(str(user_id) + "\n")
        except Exception,e:
            logs.write(e + "\n")
        
        i += 1
        print "#" + str(i), "of", array_count
    logs.write("END TIME: " + str(datetime.datetime.now()))
```
