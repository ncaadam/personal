```python
def safe_cast(val, to_type, default=0):
    try:
        return to_type(val)
    except ValueError:
        return default
```

```python
def chicago_filter(line):
    five_zip = safe_cast(line[10][0:6].replace('\"',''), int)
    if five_zip in chi_zip_list:
        return True
    else:
        return False
```

```python
def quote_filter(line):
    new_line = []
    for cell in line:
        new_line.append(cell.replace('\"',''))
    return new_line
```

```python
# create chicago zip list
chicago_zips = pd.read_csv('data/chicago_metro_zip.csv')
zip_only = chicago_zips['zip']
chi_zip_list = zip_only.tolist()
```

```python
docgraph_edge_raw = sc.textFile('../../../DocGraph/data/DocGraph-2012-2013-Days365-Edge-Public-Domain/DocGraph-2012-2013-Days365.csv').map(lambda x: x.split('\t'))
```

```python
chi_npi_list = list(docgraph_node_chicago_sf['npi'].unique())
```

```python
zip_dict = {}
for z in chi_npi_list:
    zip_dict[z] = 0
```

```python
def filter_npis(line):
    if line[0] in zip_dict and line[1] in zip_dict:
        return True
    else:
        return False
```

```python
def encode_filter(line):
    new_line = []
    for first in line:
        first = first.split(',')
        for cell in first:
            cell = cell.encode('utf-8').strip()
            new_line.append(cell)    
    return new_line
```

```python
docgraph_edge_updated = docgraph_edge_raw.map(lambda x : encode_filter(x))
```

```python
docgraph_edge_chicago = docgraph_edge_updated.filter(lambda line : filter_npis(line))
```
