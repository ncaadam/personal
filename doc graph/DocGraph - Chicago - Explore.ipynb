# Functions

```python
def safe_cast(val, to_type, default=0):
    try:
        return to_type(val)
    except ValueError:
        return default
```

```python
def chicago_filter(line):
    five_zip = safe_cast(line[10][0:6].replace('\"',''), int)
    if five_zip in chi_zip_list:
        return True
    else:
        return False
```

```python
def quote_filter(line):
    new_line = []
    for cell in line:
        new_line.append(cell.replace('\"',''))
    return new_line
```

```python
def encode_filter(line):
    new_line = []
    for first in line:
        first = first.split(',')
        for cell in first:
            cell = cell.encode('utf-8').strip()
            new_line.append(cell)    
    return new_line
```

```python
def filter_npis(line):
    if line[0] in zip_dict and line[1] in zip_dict:
        return True
    else:
        return False
```

# Provider Type Pairs - Graphs

```python
def get_provider_type_connections(nodes, edges, referee, referer):
    referee_npi_list = list(nodes.filter_by(referee,'provider_type')['npi'].unique())
    referer_npi_list = list(nodes.filter_by(referer,'provider_type')['npi'].unique())
    referee_edges = edges.filter_by(referee_npi_list,'FirstNPI')
    total_edges = referee_edges.filter_by(referer_npi_list,'SecondNPI')
    referee_nodes = nodes.filter_by(list(total_edges['FirstNPI'].unique()),'npi')
    referer_nodes = nodes.filter_by(list(total_edges['SecondNPI'].unique()),'npi')
    total_nodes = referee_nodes.append(referer_nodes)
    
    return total_nodes, total_edges
```

```python
n, e = get_provider_type_connections(docgraph_node_chicago_sf,
                                     docgraph_edge_chicago_sf,
                                     'Cardiology',
                                     'Orthopedic Surgery')
pair_g = gl.SGraph(vertices=n, edges=e, vid_field='npi', src_field='FirstNPI', dst_field='SecondNPI')
highlight = e['FirstNPI']
pair_g.show(highlight=highlight,vlabel='nppes_provider_last_org_name',vlabel_hover=True, elabel='SharedTransactionCount',elabel_hover=True)
```
