```python
from bs4 import BeautifulSoup as bs
import requests
```

```python
#DE URL
#This url should work if you change the "de" in the base url to the appropriate state for each licensee
url = 'https://sbs-de.naic.org/Lion-Web/servlet/sbsreports.LicenseeListExtController?searchState=de&searchFor=licensees&firstName=&lastName=&npnNumber=&licensenum=1341473&beName=&dbaName=&isresident=&licenseclass=&licensestatus=&loa=&city=&state=&sort=LNAME&businessType=I&x=45&y=2&courseGroup=&courseName=&providerName=&providerType=&recognizedcourse=All&providerId=&txtNaicNumber=&txtNaicGroupNumber=&txtCompName=&txtOldCompName=&hidCompanyType=&txtCompanyType=&txtDbaName=&txtOldDbaName=&txtfirstName=&txtLastName=&txtTrackingID=&txtMiddleName=&txtLicStatus=&txtBusAddressCity=&txtBusAddressState=&txtBusAddressZip=&txtMailingAddressCity=&txtMailingAddressState=&txtMailingAddressZip=&bustype='
```

```python
response = requests.get(url)
soup = bs(response.content,"html.parser")
soup
#('td',{'text':'1341473'})
```

```python
#soup.find_all('td',{'bgcolor':'d3d3d3'})
lister = []
for tr in soup.find_all('td'):
    try:
        cell_string = tr.find('a',text = True).text.strip()
        if len(cell_string) > 0:
            testing = tr.find_previous_siblings() + tr.find_next_siblings()
            for td in testing:
                if len(td.text.strip()) > 0:
                    lister.append(td.text.strip())
    except:
        continue
    tr.find_all('a',href=True)
```

```python
for item in lister:
    print item
```
