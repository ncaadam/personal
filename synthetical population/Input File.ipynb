```python
matched_records = distancesDF.idxmin(axis=1).to_dict()
```

```python
#clean up SBI file
#Fields: SBI ID, HH Income, Age, Gender, Employment Status, Liquid Asset Balance, Other Investment Balance, Home Value, Retirement Assets, Variable Annuity Balance
SBI_Input = SBI[['ID','G5','OCALC_4','OCALC_27','F10A_M','F10D_M','F10A_F','F10D_F','QX1','OCALC_3','B2_13']]
SBI_Input = SBI_Input.rename(columns = {'G5':'SBI_Home_Value','OCALC_4':'SBI_Age','OCALC_27':'SBI_Income','F10A_M':'SBI_Num_Pols_1','F10D_M':'SBI_Pol_Val_1','F10A_F':'SBI_Num_Pols_2','F10D_F':'SBI_Pol_Val_2','QX1':'Joint_Decisions','OCALC_3':'Education','B2_13':'Financial_Confidence'})
SBI_Input[['SBI_Pol_Val_1','SBI_Num_Pols_1','SBI_Pol_Val_2','SBI_Num_Pols_2','SBI_Home_Value']] = SBI_Input[['SBI_Pol_Val_1','SBI_Num_Pols_1','SBI_Pol_Val_2','SBI_Num_Pols_2','SBI_Home_Value']].fillna(0)
SBI_Input['SBI_Ann_Val'] = (SBI_Input['SBI_Pol_Val_1']+SBI_Input['SBI_Pol_Val_2'])/(SBI_Input['SBI_Num_Pols_1']+SBI_Input['SBI_Num_Pols_2'])
SBI_Input['SBI_Ann_Val'] = SBI_Input['SBI_Ann_Val'].fillna(0)
SBI_Input = SBI_Input[['ID','SBI_Home_Value','SBI_Age','SBI_Income','SBI_Ann_Val','Joint_Decisions','Education','Financial_Confidence']]
```

```python
SBI_Input.describe()
```

```python
#write function to bring in a new SBI column, rename it, and replace NAs with either 0 or the median
def add_SBI_col(col, colname, fillna):
    if fillna == 'zero':
        SBI_Input[colname] = SBI[col].fillna(0)
    elif fillna == 'median':
        SBI_Input[colname] = SBI[col].fillna(SBI[col].median())
    elif fillna == 'pad':
        SBI_Input[colname] = SBI[col].fillna(method='pad')
    #print SBI_Input[colname].describe()
```

```python
#gender
add_SBI_col('OCALC_10','SBI_Sex','pad')
#employment status
add_SBI_col('OCALC_6','SBI_Emp_Status','pad')
#Retirement Assets
add_SBI_col('FCALC_2','SBI_Ret_Assets','zero')
SBI_Input['SBI_Ret_Assets'] = SBI_Input['SBI_Ret_Assets'] - SBI_Input['SBI_Ann_Val']
#Liquid Assets
add_SBI_col('NET_3','SBI_Liq_Assets','median')
#Other Assets
add_SBI_col('NET_1','SBI_Oth_Assets','median')
SBI_Input['SBI_Oth_Assets'] = SBI_Input['SBI_Oth_Assets'] - SBI_Input['SBI_Ann_Val'] - SBI_Input['SBI_Ret_Assets'] - SBI_Input['SBI_Liq_Assets'] - SBI_Input['SBI_Home_Value']
#Mortgages
add_SBI_col('LIAB_1','SBI_Mortgage','zero')
#Mortgage Payment
add_SBI_col('G15_A','SBI_MortgagePayment','zero')
#Rent
add_SBI_col('G4C','SBI_RentPayment','zero')
#All other liabilities
add_SBI_col('NET_5','SBI_Oth_Liab','median')
SBI_Input['SBI_Oth_Liab'] = SBI_Input['SBI_Oth_Liab'] - SBI_Input['SBI_Mortgage']
SBI_Input.describe()
```

```python
#assemble input file fields from AXA and SBI datasets based on matched IDs
#Fields: SBI ID, HH Income, Age, Gender, Employment Status, Liquid Asset Balance, Other Investment Balance, Home Value, Retirement Assets, Variable Annuity Balance
#input_file = pd.DataFrame()
input_file_list = []
for axa_policy, sbi_id in matched_records.iteritems():
    matched_record = SBI_Input[SBI_Input['ID'] == sbi_id]
    matched_record['AXA_Policy_Num'] = axa_policy
    #input_file = pd.concat([input_file, matched_record])
    input_file_list.append(matched_record)
input_file = pd.concat(input_file_list)
input_file.head()
```

```python
#merge with axa data for comparison
input_file = pd.merge(input_file, AXA_Sample, how = 'left', left_on='AXA_Policy_Num', right_on='PT_POLICY_NUMBER')
input_file.head()
```

```python
#for state in data, read in region
state_region_map = pd.read_csv('/Users/sallee001/Desktop/AXA_PHB_Study/AXA Model/Input File/Updated File 9_4/State_Region_mapping.csv',sep=',')
input_file = pd.merge(input_file, state_region_map, how='left', left_on="AXA_STATE", right_on="State")
```

```python
#create cumulative wealth column
input_file['Total_Assets'] = input_file['SBI_Ret_Assets']+input_file['SBI_Liq_Assets']+input_file['SBI_Oth_Assets']+input_file['AVG_TOT_ACC_VAL']+input_file['HOME_VALUE']
#find wealth quantiles and assign values
q = pd.qcut(input_file['Total_Assets'],3,labels=["Low","Medium","High"],precision=0)
input_file['Asset Quantile'] = q
```
