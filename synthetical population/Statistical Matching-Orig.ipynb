```python
#import pandas
import pandas as pd

#import distance for matching algo
import scipy.spatial.distance as distance

from __future__ import division
```

```python
#pumd_data = pd.read_csv("../data/Interview_2013_FMLI_all.xlsx", sep=",")
pumd_excel = pd.ExcelFile("../data/Interview_2013_FMLI_all.xlsx")
```

```python
pumd_data = pumd_excel.parse()
pumd_data.head()
```

```python
#drop unnecessary columns
pumd_data_filter = pumd_data[['NEWID','SEX_REF','AGE_REF','FSALARYX','STATE','CUTENURE']]
### CONSIDER REPLACING '0' SALARIES ###
pumd_data_filter.head()
```

```python
pumd_data_filter['HOME_STATUS'] = pumd_data_filter['CUTENURE'].map({1:1,2:1,3:1,4:2,5:3,6:3})
```

```python
state_codes = pd.read_csv('../data/State Codes.csv')
```

```python
state_code_dict = dict(zip(state_codes.Code, state_codes.Abb))
```

```python
pumd_data_filter['STATE_ABB'] = pumd_data_filter['STATE'].map(state_code_dict)
```

```python
pumd_data_filter = pumd_data_filter.rename(columns={'NEWID':'PUMD_ID','SEX_REF':'PUMD_SEX','AGE_REF':'PUMD_AGE',
                                                   'FSALARYX':'PUMD_SALARY','STATE':'PUMD_STATE_CODE',
                                                   'CUTENURE':'PUMD_ORIG_HOUSE_STATUS','HOME_STATUS':'PUMD_NEW_HOME_STATUS',
                                                   'STATE_ABB':'PUMD_STATE_ABB'})
```

```python
pumd_data_final = pumd_data_filter.drop(['PUMD_STATE_CODE','PUMD_ORIG_HOUSE_STATUS'],1)
```

```python
#pull in SBI Data 
SBI = pd.read_csv('../data/SBI_Complete.csv',sep=',')
SBI.columns.values.tolist()
```

```python
SBI_Matching = SBI[['ID','OCALC_4','OCALC_10','OCALC_27','STATE','G4']]
```

```python
SBI_Matching_Final = SBI_Matching.rename(columns = {'ID':'SBI_ID','OCALC_4':'SBI_AGE','OCALC_27':'SBI_SALARY','OCALC_10':'SBI_SEX',
                                                    'STATE':'SBI_STATE','G4':'SBI_HOME_STATUS'})
```

```python
SBI_Matching.describe()
```

```python
#create distance function algorithm using Mahalanobis distance: d = (xa - xb)'*CovXX^-1*(xa-xb)
#choose number of AXA records to match and pull an array of that many records at random
#import random
#import scipy.spatial.distance as distance
#n = 20000
#AXA_Sample = axa_data_final.ix[random.sample(axa_data_final.index,n)]
#AXA_Sample.describe()
```

```python
#create matrix of distance measurements
XA = pumd_data_final[['PUMD_AGE','PUMD_SEX','PUMD_SALARY','PUMD_NEW_HOME_STATUS']]#,'PUMD_STATE_ABB']]
XB = SBI_Matching_Final[['SBI_AGE','SBI_SEX','SBI_SALARY','SBI_HOME_STATUS']]#,'SBI_STATE']]
```

```python
distances = distance.euclidean(XA,XB,metric='sqeuclidean')
```

```python
#convert distances to DF and add IDs as indices
distancesDF = pd.DataFrame(distances, index=pd.np.array(pumd_data_final[['PUMD_ID']]['PUMD_ID']))
distancesDF.columns = pd.np.array(SBI_Matching_Final[['SBI_ID']]['SBI_ID'])
distancesDF.head()
```

```python
#print distances to excel (takes alot of time, large file generated)
#distancesDF.to_csv('../data/PUMD_SBI_Distances.csv')
```

```python
matched_records = distancesDF.idxmin(axis=1).to_dict()
```

```python
#clean up SBI file
#Fields: SBI ID, HH Income, Age, Gender, Employment Status, Liquid Asset Balance, Other Investment Balance, Home Value, Retirement Assets, Variable Annuity Balance
SBI_Input = SBI[['ID','OCALC_4','OCALC_10','OCALC_27','STATE','G4']]
SBI_Input = SBI_Input.rename(columns = {'ID':'SBI_ID','OCALC_4':'SBI_AGE','OCALC_27':'SBI_SALARY','OCALC_10':'SBI_SEX',
                                        'STATE':'SBI_STATE','G4':'SBI_HOME_STATUS'})
```

```python
SBI_Input.describe()
```

```python
#assemble input file fields from AXA and SBI datasets based on matched IDs
#Fields: SBI ID, HH Income, Age, Gender, Employment Status, Liquid Asset Balance, Other Investment Balance, Home Value, Retirement Assets, Variable Annuity Balance
#input_file = pd.DataFrame()
input_file_list = []
for pumd_id, sbi_id in matched_records.iteritems():
    matched_record = SBI_Input[SBI_Input['SBI_ID'] == sbi_id]
    matched_record['PUMD_ID'] = pumd_id
    #input_file = pd.concat([input_file, matched_record])
    input_file_list.append(matched_record)
input_file = pd.concat(input_file_list)
input_file.head()
```

```python
#merge with PUMD data for comparison
input_file = pd.merge(input_file, pumd_data_final, how = 'left', left_on='PUMD_ID', right_on='PUMD_ID')
input_file
```

```python
#get default matching accuracy percentage between the two data sets
num_of_same = len(input_file[input_file.SBI_HOME_STATUS == input_file.PUMD_NEW_HOME_STATUS])
num_total = len(input_file)
(num_of_same/num_total)*100
```

```python
from sklearn import svm
from sklearn.cross_validation import train_test_split
from sklearn.metrics import confusion_matrix
```

```python
one_drop = input_file.drop('SBI_HOME_STATUS',1)
column_names = list(one_drop.columns.values)
od_train, od_test = train_test_split(one_drop, test_size=0.20, random_state = 24)
```

```python
od_train = pd.DataFrame(od_train,columns=one_drop.columns.values.tolist())
od_test = pd.DataFrame(od_test,columns=one_drop.columns.values.tolist())
```

```python
x = od_train[['SBI_AGE','SBI_SEX','SBI_SALARY','PUMD_SEX','PUMD_AGE','PUMD_SALARY']]
y = od_train['PUMD_NEW_HOME_STATUS']
```

```python
clf = svm.SVC()
clf.fit(x,y)
```

```python
test_x = od_test[['SBI_AGE','SBI_SEX','SBI_SALARY','PUMD_SEX','PUMD_AGE','PUMD_SALARY']]
od_test['HOME_STATUS_PREDICT'] = clf.predict(test_x)
```

```python
#write function to bring in a new SBI column, rename it, and replace NAs with either 0 or the median
def add_SBI_col(col, colname, fillna):
    if fillna == 'zero':
        SBI_Input[colname] = SBI[col].fillna(0)
    elif fillna == 'median':
        SBI_Input[colname] = SBI[col].fillna(SBI[col].median())
    elif fillna == 'pad':
        SBI_Input[colname] = SBI[col].fillna(method='pad')
    #print SBI_Input[colname].describe()
```

```python
#gender
add_SBI_col('OCALC_10','SBI_Sex','pad')
#employment status
add_SBI_col('OCALC_6','SBI_Emp_Status','pad')
#Retirement Assets
add_SBI_col('FCALC_2','SBI_Ret_Assets','zero')
SBI_Input['SBI_Ret_Assets'] = SBI_Input['SBI_Ret_Assets'] - SBI_Input['SBI_Ann_Val']
#Liquid Assets
add_SBI_col('NET_3','SBI_Liq_Assets','median')
#Other Assets
add_SBI_col('NET_1','SBI_Oth_Assets','median')
SBI_Input['SBI_Oth_Assets'] = SBI_Input['SBI_Oth_Assets'] - SBI_Input['SBI_Ann_Val'] - SBI_Input['SBI_Ret_Assets'] - SBI_Input['SBI_Liq_Assets'] - SBI_Input['SBI_Home_Value']
#Mortgages
add_SBI_col('LIAB_1','SBI_Mortgage','zero')
#Mortgage Payment
add_SBI_col('G15_A','SBI_MortgagePayment','zero')
#Rent
add_SBI_col('G4C','SBI_RentPayment','zero')
#All other liabilities
add_SBI_col('NET_5','SBI_Oth_Liab','median')
SBI_Input['SBI_Oth_Liab'] = SBI_Input['SBI_Oth_Liab'] - SBI_Input['SBI_Mortgage']
SBI_Input.describe()
```
