```python
#import different classification algos
from sklearn import svm,tree,neighbors
#import splitting function for splitting your dataset into training and testing
from sklearn.cross_validation import train_test_split
#import pandas for data analysis, manipulation, and viewing
import pandas as pd
```

```python
#create the excel object in pandas
excel = pd.ExcelFile("../Synthetic_AL_Input v11.xlsx")
```

```python
#parse the excel object by selecting which sheet (zero based) and identifying the row of the headers (zero based)
data = excel.parse(header=1,sheetname=2)

#get the headers in a variable (this will be used later)
headers = list(data.columns.values)
```

```python
#split the data into train and test (Train = 80%, Test = 20%)
data_train, data_test = train_test_split(data, test_size=0.20, random_state = 24)
```

```python
#put the nparrays back into Pandas DFs, add the headers
pd_train = pd.DataFrame(data_train,columns = headers)
pd_test = pd.DataFrame(data_test,columns = headers)
```

```python
#this is your dependent variable
#would be an input if it were a script
variable = 'OwnsHomeBoolean'
```

```python
#initialize your array of independent variables
x = pd_train[['Age 1','Gender 1','Income1','EmploymentStatus1','Savings1','TermImputed1','IRAImputed1','IRA Contributions 1',
              'R401kImputed1','401k Contribution 1','VAImputed1','CDImputed1','MFImputed1','CCDebt1','SSIncome1','Age 2',
              'Gender 2','Income2','EmploymentStatus2','Savings2','TermImputed2','IRAImputed2','IRA Contributions 2',
              'R401kImputed2','401k Contribution 1.1','VAImputed2','CDImputed2','MFImputed2','CCDebt2','SSIncome2']]

# initialize your dependent variable
y = pd_train[variable]
```

```python
#initialized a Support Vector Classifier
svm_model = svm.SVC()

#fit the model with our variables
svm_model.fit(x,y)
```

```python
#initialize Decision Tree Classifier
dec_tree = tree.DecisionTreeClassifier()

#fit the model with our variables
dec_tree.fit(x,y)
```

```python
#initialize an Extra Tree Classifier
x_dec_tree = tree.ExtraTreeClassifier()

#fit the model with our variables
x_dec_tree.fit(x,y)
```

```python
#initialize a K-nearest Neighbor Classifier
knn_model = neighbors.KNeighborsClassifier()

#fit the model with our variables
knn_model.fit(x,y)
```

```python
#grab our independent variables from the test dataset
test_x = pd_test[['Age 1','Gender 1','Income1','EmploymentStatus1','Savings1','TermImputed1','IRAImputed1','IRA Contributions 1',
              'R401kImputed1','401k Contribution 1','VAImputed1','CDImputed1','MFImputed1','CCDebt1','SSIncome1','Age 2',
              'Gender 2','Income2','EmploymentStatus2','Savings2','TermImputed2','IRAImputed2','IRA Contributions 2',
              'R401kImputed2','401k Contribution 1.1','VAImputed2','CDImputed2','MFImputed2','CCDebt2','SSIncome2']]

#predict our dependent variables for each of the algos we'd like to try
svm_predict = svm_model.predict(test_x)
dec_tree_predict = dec_tree.predict(test_x)
x_dec_tree_predict = x_dec_tree.predict(test_x)
knn_model_predict = knn_model.predict(test_x)
```

```python
#add the predicted columns to the test dataset with the appropriate headers
test_x[variable + '_svm_predict'] = svm_predict
test_x[variable + '_dec_tree_predict'] = dec_tree_predict
test_x[variable + '_x_dec_tree_predict'] = x_dec_tree_predict
test_x[variable + '_knn_model_predict'] = knn_model_predict
```
