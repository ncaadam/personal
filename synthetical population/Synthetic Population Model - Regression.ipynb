```python
from sklearn import cross_validation,linear_model
import pandas as pd
import numpy as np
```

```python
excel = pd.ExcelFile("../Synthetic_AL_Input v11.xlsx")
```

```python
data = excel.parse(header=1,sheetname=2)
headers = list(data.columns.values)
```

```python
data_train, data_test = cross_validation.train_test_split(data, test_size=0.20, random_state = 24)
```

```python
pd_train = pd.DataFrame(data_train,columns = headers)
pd_test = pd.DataFrame(data_test,columns = headers)
```

```python
variable = 'Num Children'
```

```python
x = pd_train[['Age 1','Gender 1','Income1','EmploymentStatus1','Savings1','TermImputed1','IRAImputed1','IRA Contributions 1',
              'R401kImputed1','401k Contribution 1','VAImputed1','CDImputed1','MFImputed1','CCDebt1','SSIncome1']]
y = pd_train[variable]
```

```python
log_reg = linear_model.LogisticRegression()
log_reg.fit(x,y)
```

```python
test_x = pd_test[['Age 1','Gender 1','Income1','EmploymentStatus1','Savings1','TermImputed1','IRAImputed1','IRA Contributions 1',
              'R401kImputed1','401k Contribution 1','VAImputed1','CDImputed1','MFImputed1','CCDebt1','SSIncome1']]
log_reg_predict = log_reg.predict(test_x)
```

```python
col_header = variable + '_log_reg_predict'
test_x[col_header] = log_reg_predict
```
